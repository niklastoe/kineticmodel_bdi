# Marginalized posterior distributions for all parameters 

Define properly formatted parameter labels

```python
var_parm_formatted = {} 

var_parm_formatted['KA'] = r'$\log(K_{\mathrm{A}} \ / \ \mathrm{M}^{-1})$'
var_parm_formatted['S0'] = r'$\log([\mathrm{S}]_0 \ / \ \mathrm{M})$'
var_parm_formatted['kd'] = r'$\log(k_\mathrm{d} \ / \ \mathrm{s}^{-1})$'
var_parm_formatted['k'] = r"$\log(k'_{\mathrm{ER}} \ / \ \mathrm{s}^{-1})$"
var_parm_formatted['sigma_ads'] = r'$\log(\sigma_{\mathrm{ads}} \ / \ \mathrm{M})$'
var_parm_formatted['sigma_kin'] = r'$\log(\sigma_{\mathrm{kin}} \ / \ \mathrm{M})$'
```

We will want a function to easily plot the (uniform) priors:

```python
def plot_prior(x_min, x_max, ax):
    margin = 0.1
    x_val = np.arange(x_min - margin, x_max + margin, 0.02)
    y_val = (x_val > x_min).astype(float) + (x_val < x_max).astype(int)
    y_val -= 1
    x_diff = (x_max - x_min)
    y_val /=  x_diff
    
    ax.plot(x_val, y_val, c='red', linestyle='--')
```

Now, plot all marginalized posterior distributions for the different levels of data inclusion ($D = \{ D_{\mathrm{KIN}} \}$, $D = \{ D_{\mathrm{KIN}}, D_{\mathrm{EQ}} \}$, $D = \{ D_{\mathrm{KIN}}, D_{\mathrm{EQ}}, D_{\mathrm{TD}} \}$).

```python
cols = 2
rows = len(true_parameteres_reduced) / cols

fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(8,9))

curr_ax = 0
parm_order = ['KA', 'S0', 'kd', 'k', 'sigma_kin', 'sigma_ads']

for idx, parm_name in enumerate(parm_order):
    x_pos = idx / cols
    y_pos = idx % cols
    axes[x_pos, y_pos].set_xlabel(var_parm_formatted[parm_name])
    
    true_val = true_parameteres_reduced[parm_name]
    axes[x_pos, y_pos].axvline(true_val, c='k')
    x_min, x_max = true_val - 2, true_val + 2
    
    for sampler in ['ER_eq_td', 'ER_eq', 'ER']:
        try:
            sns.kdeplot(samplers[sampler].parm_df[parm_name], ax=axes[x_pos, y_pos], 
                        label=sampler, legend=False, clip=(x_min, x_max))
        except KeyError: # sigma_ads is not always sampled
            pass
#             axes[x_pos, y_pos].axhline(0)
        
#     if parm_name == 'sigma_kin':
#         axes[x_pos, y_pos].legend(['KIN', 'KIN+EQ', 'KIN+EQ+TD'])
        
    

    plot_prior(x_min, x_max, axes[x_pos, y_pos])
    axes[x_pos, y_pos].set_xlim(x_min - 0.1, x_max + 0.1)
    axes[x_pos, y_pos].set_xticks([x_min, true_val, x_max])
    axes[x_pos, y_pos].set_yticks([])
    
    curr_ax += 1
    
plt.subplots_adjust(hspace=0.)
fig.text(0., 0.7, 'Probability density', rotation=90)

plt.tight_layout()
fig.savefig('/home/niklas/Documents/SAS_kinetic_modeling_BDI_manuscript/figures/proof_of_concept/parameter_space.pdf', dpi=300)
```

# Visualize difference in posterior for ER and LH model depending on $D$

Store the likelihoods for $\theta_{\mathrm{true}}$. Since we only use uniform priors, these are representative of the posteriors as well.

```python
true_parm_kin_likelihood = ER_likelihood.calc_likelihood(reformat_local_parameters(true_parameteres_reduced))
true_parm_eq_ads_likelihood = eq_adsorption_likelihood.calc_likelihood(reformat_local_parameters(true_parameteres_reduced))
true_parm_td_ads_likelihood = td_adsorption_likelihood.calc_likelihood(reformat_local_parameters(true_parameteres_reduced))

true_logp = {'KIN': true_parm_kin_likelihood,
              'EQ': true_parm_eq_ads_likelihood,
             'TD': true_parm_td_ads_likelihood}

true_logp['KIN+EQ'] = true_logp['KIN'] + true_logp['EQ']
true_logp['KIN+EQ+TD'] = true_logp['KIN'] + true_logp['EQ'] + true_logp['TD']

true_logp
```

Define a function that plots the densities for ER and LH posteriors for the same underlying $D$

```python
def plot_logp_dens(sampler, correction, ax):
    sns.kdeplot(samplers['ER' + sampler].parm_df['logp'] - true_logp[correction], ax=ax, gridsize=500, legend=False)
    sns.kdeplot(samplers['LH' + sampler].parm_df['logp'] - true_logp[correction], ax=ax, gridsize=500, legend=False)
    ax.axvline(0, c='k')
    
    ax.set_xlim(-17, 3)
    if ax != axes[2]:
        ax.set_xticks([])

    D_name = correction.split('+')
    D_names = [r'$D_{\mathrm{' + x + '}}$' for x in D_name]
    
    ax.set_title(r'$D$={' + ', '.join(D_names) + '}')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, sharex=False, sharey=True, figsize=(8,6))

plot_logp_dens('', 'KIN', axes[0])

plot_logp_dens('_eq', 'KIN+EQ', axes[1])

plot_logp_dens('_eq_td', 'KIN+EQ+TD', axes[2])

axes[0].legend([r'$M=$ER', r'$M=$LH'], framealpha=1.)

axes[1].set_ylabel(r'Probability density')
axes[1].set_yticks([])

axes[2].set_xlabel(r'$\ln \left( \frac{p(\theta|D, \ M)}{p(\theta_{\mathrm{true}} | D, \ M_{\mathrm{true}})} \right)$')

plt.tight_layout()
plt.subplots_adjust(hspace=0.5)

plt.savefig('/home/niklas/Documents/SAS_kinetic_modeling_BDI_manuscript/figures/proof_of_concept/logposterior_distribution.pdf')
```

# Posterior Predictive Check (not shown for POC in paper)

If you want to see if the inferred parameters are reasonable, use the posterior predictive check:

```python
from functools import partial
```

```python
def evaluate_rel_activity(sel_likelihood, *args, **kwargs):
    df = sel_likelihood.generate_D_from_theta(*args, **kwargs)
    df[df < 0] = np.nan
    return sel_likelihood.model.data_conversion_dict['invert']['func'](df)
```

```python
ER_rel_activity = partial(evaluate_rel_activity, ER_likelihood)
LH_rel_activity = partial(evaluate_rel_activity, LH_likelihood)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12,6))


ER_samples = samplers['ER_eq_td'].parm_df.sample(n=500)

analysis.posterior_predictive_check(ER_samples,
                                   ER_likelihood.generate_D_from_theta,
                                   axes[0][0])

analysis.posterior_predictive_check(ER_samples,
                                   td_adsorption_likelihood.generate_D_from_theta,
                                   axes[0][1])

analysis.posterior_predictive_check(ER_samples,
                                   eq_adsorption_likelihood.generate_D_from_theta,
                                   axes[0][2])


LH_samples = samplers['LH_eq_td'].parm_df.sample(n=500)

analysis.posterior_predictive_check(LH_samples,
                                   LH_likelihood.generate_D_from_theta,
                                   axes[1][0])

analysis.posterior_predictive_check(LH_samples,
                                   td_adsorption_likelihood.generate_D_from_theta,
                                   axes[1][1])

analysis.posterior_predictive_check(LH_samples,
                                   eq_adsorption_likelihood.generate_D_from_theta,
                                   axes[1][2])
```
